### Getting data for state bank of india from above symbol is SSIN.BSE

```python
import requests

url = "https://alpha-vantage.p.rapidapi.com/query"

symbol = "SBIN.BSE"

querystring = {"function":"TIME_SERIES_DAILY","symbol":symbol,"outputsize":"compact","datatype":"json"}

headers = {
	"x-rapidapi-key": "8b77d150e6mshb70f6d82abd0b55p136ad0jsne10f02d546f2",
	"x-rapidapi-host": "alpha-vantage.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

data = response.json()
```

```python
data
```

```python
data.keys()
```

```python
data['Time Series (Daily)']
```

```python
import pandas as pd 

df = pd.DataFrame(data['Time Series (Daily)']).T

df.head()
```

```python
df.info()
```

```python
df.index
```

### Convert index to datetime

```python
df.index = pd.to_datetime(df.index)
```

```python
df.index
```

### Convert all data into float

```python
for i in df.columns:
    df[i] = df[i].astype(float)
```

```python
df
```

```python
df.index.name = "Date"
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.to_csv("SBIStockData.csv")
```
